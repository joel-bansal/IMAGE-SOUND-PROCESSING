# TRAINING THE MODEL

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.yaml")  # build a new model from scratch

results = model.train(data="/kaggle/input/car-in-yaml/data.yaml",epochs=75)  # train the model
```

# USING EASY OCR

```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO
import easyocr

cropped_image = []
def number_plates(path):
    results = model(path)
    for r in results: 
        im_array = r.plot()  # plot a BGR numpy array of predictions
        im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
        plt.imshow(im)
        plt.show()
        coordinates = r.boxes.xyxy.tolist() #extract coordinates of the detected block
        if len(coordinates)>0:
            x1, y1, x2, y2 = coordinates[0]
            cropped_image = im_array[int(y1):int(y2), int(x1):int(x2)]
            img1 = cv2.GaussianBlur(cropped_image,(9,9),0) #reduced noise
            edge = cv2.Canny(img1,50,100)
            cropped_image = cv2.cvtColor(cropped_image,cv2.COLOR_BGR2GRAY)
            reader = easyocr.Reader(['en'])
            result = reader.readtext(cropped_image, detail=0)
            plt.imshow(cropped_image,cmap = 'gray')
            plt.show()
            print("Number Plate : ", result)
        else:
            print("No detection")
# number_plates('/kaggle/input/car-in-yaml/test/images/Cars430_png.rf.caae106c8c02928f8ba8eba3023b494f.jpg')
```
